```python
# change the row word
#school column 
mat['school'] = mat['school'].replace('GP', 'Gabriel Pereira')
mat['school'] = mat['school'].replace('MS', 'Mousinho da Silveira')

#sex
mat['sex'] = mat['sex'].replace('F', 'Female')
mat['sex'] = mat['sex'].replace('M', 'Male')

#address
mat['address'] = mat['address'].replace('U', 'Urban')
mat['address'] = mat['address'].replace('R', 'Rural')

#famsize
mat['famsize'] = mat['famsize'].replace('LE3', 'equal to 3')
mat['famsize'] = mat['famsize'].replace('GT3', 'greater than 3')


#Pstatus
mat['Pstatus'] = mat['Pstatus'].replace('T', 'living together')
mat['Pstatus'] = mat['Pstatus'].replace('A', 'Apart')
```

```python
#Medu
mat['Medu'] = mat['Medu'].replace('0', 'none')
mat['Medu'] = mat['Medu'].replace('1', 'primary education (4th grade)')
mat['Medu'] = mat['Medu'].replace('2', '5th to 9th grade')
mat['Medu'] = mat['Medu'].replace('3', 'secondary education')
mat['Medu'] = mat['Medu'].replace('4', 'higher education')

#Fedu
mat['Fedu'] = mat['Fedu'].replace('0', 'none')
mat['Fedu'] = mat['Fedu'].replace('1', 'primary education (4th grade)')
mat['Fedu'] = mat['Fedu'].replace('2', '5th to 9th grade')
mat['Fedu'] = mat['Fedu'].replace('3', 'secondary education')
mat['Fedu'] = mat['Fedu'].replace('4', 'higher education')



#traveltime
mat['traveltime'] = mat['traveltime'].replace('1', '<15 min')
mat['traveltime'] = mat['traveltime'].replace('2', '15 to 30 min.')
mat['traveltime'] = mat['traveltime'].replace('3', '30 min. to 1 hour')
mat['traveltime'] = mat['traveltime'].replace('4', ' >1 hour')

#studytime
mat['studytime'] = mat['studytime'].replace('1', '<2 hours')
mat['studytime'] = mat['studytime'].replace('2', '2 to 5 hours')
mat['studytime'] = mat['studytime'].replace('3', '5 to 10 hours')
mat['studytime'] = mat['studytime'].replace('4', '  >10 hours')



#famrel
mat['famrel'] = mat['famrel'].replace('1', 'Very bad')
mat['famrel'] = mat['famrel'].replace('2', 'Bad')
mat['famrel'] = mat['famrel'].replace('3', 'Average')
mat['famrel'] = mat['famrel'].replace('4', 'Good')
mat['famrel'] = mat['famrel'].replace('5', 'Excellent')


#freetime 
mat['freetime'] = mat['freetime'].replace('1', 'Very Low')
mat['freetime'] = mat['freetime'].replace('2', 'Low')
mat['freetime'] = mat['freetime'].replace('3', 'Medium')
mat['freetime'] = mat['freetime'].replace('4', 'High')
mat['freetime'] = mat['freetime'].replace('5', 'Very High')


#goout 
mat['goout'] = mat['goout'].replace('1', 'Very Low')
mat['goout'] = mat['goout'].replace('2', 'Low')
mat['goout'] = mat['goout'].replace('3', 'Medium')
mat['goout'] = mat['goout'].replace('4', 'High')
mat['goout'] = mat['goout'].replace('5', 'Very High')


#Dalc 
mat['Dalc'] = mat['Dalc'].replace('1', 'Very Low')
mat['Dalc'] = mat['Dalc'].replace('2', 'Low')
mat['Dalc'] = mat['Dalc'].replace('3', 'Medium')
mat['Dalc'] = mat['Dalc'].replace('4', 'High')
mat['Dalc'] = mat['Dalc'].replace('5', 'Very High')

#Walc 
mat['Walc'] = mat['Walc'].replace('1', 'Very Low')
mat['Walc'] = mat['Walc'].replace('2', 'Low')
mat['Walc'] = mat['Walc'].replace('3', 'Medium')
mat['Walc'] = mat['Walc'].replace('4', 'High')
mat['Walc'] = mat['Walc'].replace('5', 'Very High')

#health
mat['health'] = mat['health'].replace('1', 'Very Low')
mat['health'] = mat['health'].replace('2', 'Low')
mat['health'] = mat['health'].replace('3', 'Medium')
mat['health'] = mat['health'].replace('4', 'High')
mat['health'] = mat['health'].replace('5', 'Very High')

mat['Medu'] = mat['Medu'].astype(str)
mat['Fedu'] = mat['Fedu'].astype(str)
mat['traveltime'] = mat['traveltime'].astype(str)
mat['studytime'] = mat['studytime'].astype(str)
mat['famrel'] = mat['famrel'].astype(str)
mat['freetime'] = mat['freetime'].astype(str)
mat['goout'] = mat['goout'].astype(str)
mat['Dalc'] = mat['Dalc'].astype(str)
mat['Walc'] = mat['Walc'].astype(str)
mat['health'] = mat['health'].astype(str)
```
