```python
#Delete row with null value from order_date_time
cu = cu.dropna(subset=['order_date_time'], axis=0)

#to drop the column connected_handling_time , it is 99% emty 
cu.drop(columns=['connected_handling_time'], inplace=True)

#to drop the column (Issue_reported at) and (Survey_response_Date) , it is same value of date of column (issue_responded)
cu.drop(columns=['Issue_reported at'], inplace=True)
cu.drop(columns=['Survey_response_Date'], inplace=True)

#to drop the column Customer Remarks , it is comment (not need)
cu.drop(columns=['Customer Remarks'], inplace=True)
```

```python
#change a nameof column to lower case 
cu.columns = cu.columns.str.lower()
```

```python
# change a  object for column ('order_date_time' and 'issue_responded') to the datetime
cu['order_date_time'] = pd.to_datetime(cu['order_date_time'])
```

```python
cu['issue_responded'] = pd.to_datetime(cu['issue_responded'], format='%d/%m/%Y %H:%M')
```

```python
# rename a column (agent_name and agent shift ) to (employee name and shift)
cu = cu.rename(columns={'agent_name': 'employee_name'})
cu = cu.rename(columns={'agent shift': 'employee shift'})
cu = cu.rename(columns={'category': 'problems'})
cu = cu.rename(columns={'sub-category': 'sub-probkems'})
```

```python
# add a "$" for item_price
cu['item_price'] = cu['item_price'].apply(lambda x: '${:}'.format(x))
```

```python
#add column for age_in_days
cu['age_daytime'] = (cu['issue_responded'] - cu['order_date_time']).dt.days
```
